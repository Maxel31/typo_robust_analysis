# quant_typo_neuron 分析
量子化 x typo 頑健性の実験結果を可視化する。

```python
from pathlib import Path

from quant_typo_neuron.analysis.aggregate import aggregate_results
from quant_typo_neuron.analysis.plot_results import (
    plot_accuracy_heatmap,
    plot_robustness_comparison,
    plot_perplexity_bars,
)

RESULTS_ROOT = Path("../../results")
df = aggregate_results(RESULTS_ROOT)
print(f"Total results: {len(df)}")
df.head()
```

```python
# Accuracy heatmap (clean condition only)
clean_df = df[df["typo_type"] == "clean"] if "typo_type" in df.columns else df
if not clean_df.empty:
    plot_accuracy_heatmap(clean_df)
```

```python
# Robustness comparison across quantization methods
if not df.empty:
    plot_robustness_comparison(df)
```

```python
# Perplexity comparison
if not df.empty:
    plot_perplexity_bars(df)
```

```python
# Export aggregated results
if not df.empty:
    output = Path("../../results/aggregated.csv")
    df.to_csv(output, index=False)
    print(f"Saved to {output}")
```
