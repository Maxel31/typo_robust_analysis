# 分析・可視化
`results/` に保存した結果を読み込み、頑健性カーブを描く。

```python
import json
from pathlib import Path
from typo_utils.viz import plot_robustness_curve

# sweep.py の出力（results/<exp>_sweep/sweep/predictions.jsonl）を読み込む
rows = [json.loads(l) for l in Path('results/repro_baseline_sweep/sweep/predictions.jsonl').read_text().splitlines() if l.strip()]
rates = [r['typo_rate'] for r in rows]
accs = [r['accuracy'] for r in rows]
plot_robustness_curve(rates, accs, label='baseline', save_path='analysis/figures/robustness_curve.png')
```
